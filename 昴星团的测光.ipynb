# 昴星团的光电测光实验
## 实验原理与方法
+ 利用观测星等和绝对星等之间的关系，计算得到星系离我们的距离D：$$lgD=\frac{m-M+5}{5}$$，其中D的单位是pc
+ 我们利用UV波段的测光，即可以测出恒星的色指数（温度）。赫罗图给出了主序星绝对星等和温度之间的关系，由此，我们可以得到恒星的绝对星等。
+ 得到了绝对星等之后，加上我们可以直接测量的观测星等，即由以上提到的公式计算昴星团中恒星距我们的距离。
### 下面简单介绍一下赫罗图（H-R diagram)
+ 赫罗图是以将恒星的绝对星等（光度）相对于光谱类型（有效温度）绘制的散布图。更简单的说，它将每颗恒星绘制在一张图表上，可以测量它的温度（颜色）和光度，而它与每颗恒星的位置无关。它是由Hertzsprung和Russell在20世纪10年代创立，让我们对恒星的演化过程有了更加深入的理解。
- 赫罗图可显示恒星的演化过程，大约90%的恒星位于赫罗图左上角至右下角的带状上，这条线称为主序带。位于主序带上的恒星称为主序星。形成恒星的分子云是位于图中极右的区域，但随着分子云开始收缩，其温度开始上升，会慢慢移向主序带。恒星临终时会离开主序带，恒星会往右上方移动，这里是红巨星及红超巨星的区域，都是表面温度低而光度高的恒星。经过红巨星但未发生超新星爆炸的恒星会越过主序带移向左下方，这里是表面温度高而光度低的区域，是白矮星的所在区域，接着会因为能量的损失，渐渐变暗成为黑矮星。
+ 在本实验中，我们使用颜色-星等图（colour–magnitude diagram）绘制的是恒星的视星等和颜色，通常是针对恒星都在相同距离上的星团绘制。
## 实验步骤
+ 登陆系统，选择一个合适的望远镜，并打开望远镜的控制台；
+ 按下“tracking”键，让视野停止移动，用“slew”键找到合适的观测对象，从“Finder”模式切换成“Telescope”模式，进行测光操作；
+ 找到一片没有星光的区域，进行“Sky reading”，即记录背景天光。将"integration time" 调到10.0（默认1.0），将"# of integrations"调到5（默认3）。"Start"后，分别记录下B和V波段下的mean sky counts(counts/sec)。
+ 对昴星团中的恒星进行测光操作，即"Star reading"。对于一般的源，将"integration time"调到1.0，"# of integration"调到5，保证S/N ratio 在100以上；对于特别暗的源。增加"integration time"，使得S/N ratio至少在70 以上。分别对B，V波段重复以上操作，记录下实验数据。
+ 系统会自动画出由V apparent mag vs B-V 的赫罗图，然后我们就可以找到一条主序带。系统还给出了绝对星等空间下的赫罗图，将其中的主序带向下平移，与原主序带重合，记录下平移的值
